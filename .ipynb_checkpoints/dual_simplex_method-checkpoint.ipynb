```python
import numpy as np

m = 3
n = 8
c = np.array([5., 2., 3., -16., 1., 3., -3., -12.])
b = np.array([-2., -4., -2.])
A = np.array([[-2., -1., 1., -7., 0., 0., 0., 2.],
              [4., 2., 1., 0., 1., 5., -1.,-5.],
              [1., 1., 0., -1., 0., 3., -1., 1.],])
J_B = np.array([1, 2, 3])
y_0 = np.array([1., 2., -1.])
```

```python
def dual_simplex_method(A, b, c, J_B, y_0, delta, m, n):
    iteration = 1
    J_B = J_B - 1
    x_0 = [0 for _ in range(n)]
    
    while True:
        print("Iteration #", iteration)
        J_N = np.delete(np.arange(n), J_B)
        A_B = A[:, J_B]
        print("Basis A matrix: ", np.round(A_B, 3))
        B = np.linalg.inv(A_B)
        print("Inverse A matrix B: ", B)
        Chi = B.dot(b)
        print("Pseudo−plan Chi (Bb): ", np.round(Chi, 3))
        if(min(Chi) >= 0):
            for j, chi in zip(J_B, Chi):
                x_0[j] = chi
            print("All Chi components is more or eq to zero, so wefind the optimal plan")
            print(np.round(x_0, 3), "−Optimal plan")
            print("Max profit: ", np.round(c.dot(x_0), 3))
            print("Iterations: ", iteration)
            break
        else:
            k = np.argmin(Chi)
            j_k = J_B[k]
            print("Basis index j_k: ", j_k)
            my = B[k].dot(A)
            print("Mu values: ", my)
            sigma = []
            for j_n in J_N:
                if my[j_n] >= 0:
                    sigma.append(np.inf)
                else:
                    sigma.append(- delta[j_n] / my[j_n])
            print("Sigmas value: ", np.round(sigma, 3))
            sigma_0 = min(sigma)
            print("Minimal sigma is: ", np.round(sigma_0, 3))
            sigma_j = J_N[np.argmin(sigma)]
            print("Minimal sigma’s index is: ", sigma_j)
                
            if sigma_0 == np.inf:
                print("The primal problem’s constraints areincompatible")
                break 
            else:
                print("Build new dual plan and ’coplan ’...")
                y_0 = y_0 + sigma_0 * B[k]
                delta = delta + sigma_0 * my
                J_B[k] = sigma_j
                print("New dual plan: ", np.round(y_0, 3))
                print("New ’coplan’: ", np.round(delta, 3))
                print("New basis indexes: ", J_B + 1)
                print("\n")
                iteration = iteration + 1 
                
def find_coplan(y_0, A, c):
    return y_0 @ A - c
```

```python
dual_simplex_method(A, b, c, J_B, y_0, find_coplan(y_0, A, c), m, n)
```
