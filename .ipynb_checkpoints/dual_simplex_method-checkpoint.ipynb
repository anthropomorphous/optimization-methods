```python
import numpy as np

c = np.array([5., 2., 3., -16., 1., 3., -3., -12.])
b = np.array([-2., -4., -2.])
matrix_A = np.array([[-2., -1., 1., -7., 0., 0., 0., 2.],
              [4., 2., 1., 0., 1., 5., -1.,-5.],
              [1., 1., 0., -1., 0., 3., -1., 1.],])
initial_j_b = np.array([1, 2, 3])
y_0 = np.array([1., 2., -1.])
```

```python
def dual_simplex_method(matrix_A, b, c, initial_j_b, y_0):
    n = len(c)
    initial_j_b = initial_j_b - 1
    x_0 = [0 for _ in range(n)]
    delta = find_coplan(y_0, matrix_A, c)
    i = 1
    
    while True:
        j_n = np.delete(np.arange(n), initial_j_b)
        basis_matrix = matrix_A[:, initial_j_b]
        inverse_basis_matrix = np.linalg.inv(basis_matrix)
        pseudo_plan = inverse_basis_matrix @ b
        print("Итерация ", i, "\nА: ", basis_matrix, "\nB: ", inverse_basis_matrix, "\nПсевдоплан: ", np.round(pseudo_plan, 4))
        
        if(min(pseudo_plan) >= 0):
            for index, value in zip(initial_j_b, pseudo_plan):
                x_0[index] = value
            print("\nОптимальный план", np.round(x_0, 4), "\nИтераций: ", i)
            break
        k = np.argmin(pseudo_plan)
        j_k = initial_j_b[k]
        print("Basis index j_k: ", j_k)
        my = inverse_basis_matrix[k] @ matrix_A
        print("Mu values: ", my)
        sigma = []
        for index in j_n:
            if my[index] >= 0:
                sigma.append(np.inf)
            else:
                sigma.append(- delta[index] / my[index])
        print("Sigmas value: ", np.round(sigma, 4))
        sigma_0 = min(sigma)
        print("Minimal sigma is: ", np.round(sigma_0, 4))
        sigma_j = j_n[np.argmin(sigma)]
        print("Minimal sigma’s index is: ", sigma_j)
            
        negative_elements_indexes = get_negative_elements_indexes(pseudo_plan)
        if len(negative_elements_indexes) == 0:
            break

        y_0 = y_0 + sigma_0 * inverse_basis_matrix[k]
        delta = delta + sigma_0 * my
        initial_j_b[k] = sigma_j
        print("Новый двойственный план: ", np.round(y_0, 4), "\nНовый коплан: ", np.round(delta, 4), "\nНовый базис: ", initial_j_b + 1)
        i += 1 
                
def find_coplan(y_0, A, c):
    return y_0 @ A - c

def get_negative_elements_indexes(elements_list):
    return [index for index, value in enumerate(elements_list) if value < 0]
```

```python
dual_simplex_method(matrix_A, b, c, initial_j_b, y_0)
```
