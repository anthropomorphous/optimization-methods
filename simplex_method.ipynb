```python
def simplex_main_phaze(c, a, b, x, j, m, n):
    i = 1
    j -= 1
    
    while True:
        A_t = a[:, j]
        B = np.linalg.inv(A_t)
        c_b = c[j]
        u = c_b @ B
        j_inv = delete(arange(n), j)
        delta = list(map(lambda inv: u @ a[:, inv] - c[inv], list(j_inv)))
        theta = []
        
        print("\nA: ", A_t, "\nB: ", B, "\nc_b: ", c_b, "\nВектор потенциалов: ", u, "\nВектор оценок: ", delta)
        
        j0 = j_inv[argmin(delta)]
        
        if is_optimal_solution(delta):
            print("\n\nРезультат: \nОптимальный план x0: ", list(map(lambda value: round(float(value), 4), list(x))), 
                  "\nc'x0: ", round(float(c @ x), 4), "\nКоличество итераций: ", i)
            return list(map(lambda value: round(float(value), 4), list(x))), round(float(c @ x), 4), i
            break
        
        z = B @ a[:, j0]
        print('z: ', z)
        
        if not all(value <= 0 for value in list(z)):
            print("Нет решений")
            break
                
        theta = [x[j[index]] / value if value > 0 else math.inf for index, value in enumerate(z)]
        index_min, theta_min = min(enumerate(theta), key=lambda pair: pair[1])
                
        s = index_min   
        print("s:", s + 1, "\njs: ", j[s] + 1)
        
        x[j_inv] = 0
        x[j0] = theta_min
        
        for index, value in zip(j, z):
            x[index] -= theta_min * value
            
        j[s] = j0
        print("Новый базисный план x: ", x, "\nНовый базис: ", j+1, "\n")
        i += 1
        
        
def is_optimal_solution(delta):
    for index, value in enumerate(delta):
        if value < 0:
            return False
    return True


simplex_main_phaze(c, A, b, x, Jb, m, n)
```
